### 1.5 Implement Key Expansion

```python
# Round constant words
RCON = [0x01, 0x02, 0x04, 0x08, 0x10, 0x20, 0x40, 0x80, 0x1b, 0x36]

def xtime(a):
    """Galois Field multiplication by 2 (xtime operation)"""
    return (((a << 1) ^ 0x1B) & 0xFF) if (a & 0x80) else (a << 1)

def multiply(a, b):
    """Galois Field multiplication of a and b"""
    return reduce(lambda x, y: x ^ y, [((b >> i) & 1) * a for i in range(8)])

def pad(data, block_size=16):
    """PKCS7 padding"""
    padding_len = block_size - (len(data) % block_size)
    return data + bytes([padding_len] * padding_len)

def unpad(data):
    """Remove PKCS7 padding"""
    return data[:-data[-1]]
```

```python
def key_expansion(key):
    """Expand the key into a key schedule"""
    key_words = [key[i:i+4] for i in range(0, len(key), 4)]
    for i in range(4, 44):
        word = list(key_words[i-1])
        if i % 4 == 0:
            word = [SBOX[word[1]] ^ RCON[i//4-1], SBOX[word[2]], SBOX[word[3]], SBOX[word[0]]]
        key_words.append([a ^ b for a, b in zip(key_words[i-4], word)])
    return [byte for word in key_words for byte in word]

# Test Key Expansion
test_key = bytes.fromhex('2b7e151628aed2a6abf7158809cf4f3c')
expanded_key = key_expansion(test_key)
print("Original key:", test_key.hex())
print("Expanded key:", bytes(expanded_key).hex())
```

## 2. AES Encryption and Decryption

Now that we have all the building blocks, let's implement the full AES encryption and decryption processes.

### 2.1 Implement AddRoundKey


```python
def add_round_key(state, round_key):
    """AddRoundKey step: XOR the state with the round key"""
    return [[state[i][j] ^ round_key[i*4 + j] for j in range(4)] for i in range(4)]
```

### 2.2 Implement AES Encryption

```python
def aes_encrypt(plaintext, key):
    """Encrypt a 16-byte plaintext block using AES"""
    state = [list(plaintext[i:i+4]) for i in range(0, 16, 4)]
    key_schedule = key_expansion(key)
    
    state = add_round_key(state, key_schedule[:16])
    
    for round in range(1, 10):
        state = sub_bytes(state)
        state = shift_rows(state)
        state = mix_columns(state)
        state = add_round_key(state, key_schedule[round*16:(round+1)*16])
    
    state = sub_bytes(state)
    state = shift_rows(state)
    state = add_round_key(state, key_schedule[160:])
    
    return bytes(sum(state, []))
```

```python
### 2.3 Implement AES Decryption

def aes_decrypt(ciphertext, key):
    """Decrypt a 16-byte ciphertext block using AES"""
    state = [list(ciphertext[i:i+4]) for i in range(0, 16, 4)]
    key_schedule = key_expansion(key)
    
    state = add_round_key(state, key_schedule[160:])
    
    for round in range(9, 0, -1):
        state = inv_shift_rows(state)
        state = inv_sub_bytes(state)
        state = add_round_key(state, key_schedule[round*16:(round+1)*16])
        state = inv_mix_columns(state)
    
    state = inv_shift_rows(state)
    state = inv_sub_bytes(state)
    state = add_round_key(state, key_schedule[:16])
    
    return bytes(sum(state, []))

# Test AES Decryption
decrypted = aes_decrypt(ciphertext, test_key)
print("Decrypted:", decrypted)
```

## 3. Implement Different Modes of Operation

Now let's implement different modes of operation: ECB, CBC, CFB, and OFB.

```python
### 3.1 Electronic Codebook (ECB) Mode

def aes_ecb_encrypt(plaintext, key):
    """Encrypt plaintext using AES in ECB mode"""
    plaintext = pad(plaintext)
    return b''.join(aes_encrypt(plaintext[i:i+16], key) for i in range(0, len(plaintext), 16))

def aes_ecb_decrypt(ciphertext, key):
    """Decrypt ciphertext using AES in ECB mode"""
    return unpad(b''.join(aes_decrypt(ciphertext[i:i+16], key) for i in range(0, len(ciphertext), 16)))

# Test ECB mode
ecb_ciphertext = aes_ecb_encrypt(b'This is a test message for ECB mode.', test_key)
ecb_decrypted = aes_ecb_decrypt(ecb_ciphertext, test_key)
print("ECB Ciphertext:", ecb_ciphertext.hex())
print("ECB Decrypted:", ecb_decrypted)
```

```python
### 3.2 Cipher Block Chaining (CBC) Mode


def aes_cbc_encrypt(plaintext, key, iv):
    """Encrypt plaintext using AES in CBC mode"""
    plaintext = pad(plaintext)
    blocks = [plaintext[i:i+16] for i in range(0, len(plaintext), 16)]
    ciphertext = b''
    prev = iv
    for block in blocks:
        xored = bytes(a ^ b for a, b in zip(block, prev))
        encrypted_block = aes_encrypt(xored, key)
        ciphertext += encrypted_block
        prev = encrypted_block
    return ciphertext

def aes_cbc_decrypt(ciphertext, key, iv):
    """Decrypt ciphertext using AES in CBC mode"""
    blocks = [ciphertext[i:i+16] for i in range(0, len(ciphertext), 16)]
    plaintext = b''
    prev = iv
    for block in blocks:
        decrypted_block = aes_decrypt(block, key)
        plaintext += bytes(a ^ b for a, b in zip(decrypted_block, prev))
        prev = block
    return unpad(plaintext)

# Test CBC mode
iv = b'1234567890123456'
cbc_ciphertext = aes_cbc_encrypt(b'This is a test message for CBC mode.', test_key, iv)
cbc_decrypted = aes_cbc_decrypt(cbc_ciphertext, test_key, iv)
print("CBC Ciphertext:", cbc_ciphertext.hex())
print("CBC Decrypted:", cbc_decrypted)

```

```python

### 3.3 Cipher Feedback (CFB) Mode


def aes_cfb_encrypt(plaintext, key, iv):
    """Encrypt plaintext using AES in CFB mode"""
    blocks = [plaintext[i:i+16] for i in range(0, len(plaintext), 16)]
    ciphertext = b''
    prev = iv
    for block in blocks:
        encrypted_iv = aes_encrypt(prev, key)
        xored = bytes(a ^ b for a, b in zip(block, encrypted_iv))
        ciphertext += xored
        prev = xored
    return ciphertext

def aes_cfb_decrypt(ciphertext, key, iv):
    """Decrypt ciphertext using AES in CFB mode"""
    blocks = [ciphertext[i:i+16] for i in range(0, len(ciphertext), 16)]
    plaintext = b''
    prev = iv
    for block in blocks:
        encrypted_iv = aes_encrypt(prev, key)
        xored = bytes(a ^ b for a, b in zip(block, encrypted_iv))
        plaintext += xored
        prev = block
    return plaintext

# Test CFB mode
cfb_ciphertext = aes_cfb_encrypt(b'This is a test message for CFB mode encryption.', test_key, iv)
cfb_decrypted = aes_cfb_decrypt(cfb_ciphertext, test_key, iv)
print("CFB Ciphertext:", cfb_ciphertext.hex())
print("CFB Decrypted:", cfb_decrypted)

```

# RSA

RSA (Rivest-Shamir-Adleman) is an asymmetric cryptographic algorithm used for secure data transmission.
It involves a pair of keys: a public key for encryption and a private key for decryption.

```python
## Step 2: Prime Number Generation
# RSA encryption starts with generating two large prime numbers.
# These prime numbers will be used to generate the public and private keys.

def generate_prime_candidate(length):
    # Generate a random odd integer of specified bit length
    p = random.getrandbits(length)
    # Apply a mask to set MSB and LSB to 1 to ensure the number is of the required bit length and odd
    p |= (1 << length - 1) | 1
    return p

def generate_prime_number(length=1024):
    # Keep generating random odd integers until a prime number is found
    p = 4
    while not isprime(p):
        p = generate_prime_candidate(length)
    return p

# Generate two large prime numbers p and q
p = generate_prime_number()
q = generate_prime_number()

print(f"Prime number p: {p}")
print(f"Prime number q: {q}")

## Step 3: Compute n and φ(n)
# n is the product of the two primes p and q.
# φ(n) (Euler's totient function) is calculated as (p-1)*(q-1).

n = p * q
phi_n = (p - 1) * (q - 1)

print(f"n (p * q): {n}")
print(f"φ(n) ((p-1)*(q-1)): {phi_n}")

## Step 4: Choose Public Key e
# Choose an integer e such that 1 < e < φ(n) and gcd(e, φ(n)) = 1.
# Commonly used value for e is 65537.

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

e = 65537
if gcd(e, phi_n) != 1:
    raise ValueError("e and φ(n) are not coprime")

print(f"Public key exponent e: {e}")

## Step 5: Compute Private Key d
# Compute the modular multiplicative inverse of e modulo φ(n), which is d.
# This can be done using the Extended Euclidean Algorithm.

d = mod_inverse(e, phi_n)
print(f"Private key exponent d: {d}")

## Step 6: Key Pair
# The public key is (e, n) and the private key is (d, n).

public_key = (e, n)
private_key = (d, n)

print(f"Public key: {public_key}")
print(f"Private key: {private_key}")

## Step 7: Encryption
# To encrypt a message, we convert the message into an integer m such that 0 <= m < n.
# The ciphertext c is calculated as c = m^e mod n.

def rsa_encrypt(plaintext, public_key):
    e, n = public_key
    m = int.from_bytes(plaintext.encode(), 'big')
    if m >= n:
        raise ValueError("Message too large for the key size")
    c = pow(m, e, n)
    return c

# Encrypt a sample message
plaintext = "Hello, RSA!"
ciphertext = rsa_encrypt(plaintext, public_key)
print(f"Ciphertext: {ciphertext}")

## Step 8: Decryption
# To decrypt a ciphertext, we compute m = c^d mod n.
# We then convert the integer m back to a plaintext message.

def rsa_decrypt(ciphertext, private_key):
    d, n = private_key
    m = pow(ciphertext, d, n)
    plaintext_bytes = m.to_bytes((m.bit_length() + 7) // 8, 'big')
    return plaintext_bytes.decode()

# Decrypt the sample message
decrypted_message = rsa_decrypt(ciphertext, private_key)
print(f"Decrypted message: {decrypted_message}")
```

# Hash Functions

```python
# MD5 Hash Function Implementation from Scratch

## Step 1: Introduction to MD5
# MD5 (Message-Digest Algorithm 5) is a widely used hash function that produces a 128-bit hash value.

# Step 2: Helper Functions and Constants
# MD5 involves several helper functions and constants. We start by defining these.

# Import necessary libraries
import struct

# Constants for MD5
s = [7, 12, 17, 22] * 4 + [5, 9, 14, 20] * 4 + [4, 11, 16, 23] * 4 + [6, 10, 15, 21] * 4
K = [int(abs(struct.unpack("i", struct.pack("f", i))[0])) for i in range(1, 65)]

# F, G, H, and I are non-linear functions used in the algorithm
def F(x, y, z): return (x & y) | (~x & z)
def G(x, y, z): return (x & z) | (y & ~z)
def H(x, y, z): return x ^ y ^ z
def I(x, y, z): return y ^ (x | ~z)

# Rotate left function
def left_rotate(x, c):
    return (x << c) | (x >> (32 - c))

# Step 3: Preprocessing
# Padding the message so its length in bits is congruent to 448 (mod 512).
def md5_preprocess(message):
    original_byte_len = len(message)
    original_bit_len = original_byte_len * 8
    
    # Append the bit '1' to the message
    message += b'\x80'
    
    # Append 0 <= k < 512 bits '0', such that the resulting message length (in bits)
    # is congruent to 448 (mod 512)
    while len(message) % 64 != 56:
        message += b'\x00'
    
    # Append the original message length as a 64-bit little-endian integer
    message += struct.pack('<Q', original_bit_len)
    
    return message

# Step 4: Processing Message in 512-bit chunks
# MD5 processes the message in successive 512-bit chunks.
def md5_process_chunk(chunk, h0, h1, h2, h3):
    a, b, c, d = h0, h1, h2, h3
    
    for i in range(64):
        if 0 <= i <= 15:
            f = F(b, c, d)
            g = i
        elif 16 <= i <= 31:
            f = G(b, c, d)
            g = (5 * i + 1) % 16
        elif 32 <= i <= 47:
            f = H(b, c, d)
            g = (3 * i + 5) % 16
        elif 48 <= i <= 63:
            f = I(b, c, d)
            g = (7 * i) % 16
        
        f = (f + a + K[i] + struct.unpack('<I', chunk[4 * g:4 * g + 4])[0]) & 0xFFFFFFFF
        a, d, c, b = d, (b + left_rotate(f, s[i])) & 0xFFFFFFFF, b, c
    
    h0 = (h0 + a) & 0xFFFFFFFF
    h1 = (h1 + b) & 0xFFFFFFFF
    h2 = (h2 + c) & 0xFFFFFFFF
    h3 = (h3 + d) & 0xFFFFFFFF
    
    return h0, h1, h2, h3

# Step 5: MD5 Hash Function
# The main MD5 function that ties everything together.
def md5(message):
    message = md5_preprocess(message)
    
    # Initialize variables:
    h0, h1, h2, h3 = 0x67452301, 0xEFCDAB89, 0x98BADCFE, 0x10325476
    
    # Process the message in successive 512-bit chunks:
    for i in range(0, len(message), 64):
        h0, h1, h2, h3 = md5_process_chunk(message[i:i+64], h0, h1, h2, h3)
    
    # Produce the final hash value (digest)
    return struct.pack('<4I', h0, h1, h2, h3).hex()

# Example usage
message = "Hello, MD5!"
hash_value = md5(message.encode())
print(f"MD5 Hash: {hash_value}")
```

# HMAC

```python
# Function to implement HMAC using MD5
def hmac_md5(key, message):
    # Block size for MD5 is 64 bytes
    block_size = 64

    # If key is longer than block size, hash it
    if len(key) > block_size:
        key = bytes.fromhex(md5(key))
    
    # If key is shorter than block size, pad it with zeros
    if len(key) < block_size:
        key = key.ljust(block_size, b'\x00')
    
    # Create inner and outer padded keys
    o_key_pad = bytearray((x ^ 0x5C) for x in key)
    i_key_pad = bytearray((x ^ 0x36) for x in key)

    # Perform inner MD5
    inner_md5 = md5(i_key_pad + message)
    
    # Perform outer MD5
    hmac_result = md5(o_key_pad + bytes.fromhex(inner_md5))
    
    return hmac_result

# Example usage
key = b"secret_key"
message = b"Hello, HMAC!"
hmac_value = hmac_md5(key, message)
print(f"HMAC-MD5: {hmac_value}")
```

# Diffie-Hellman Key Exchange

```python
# Diffie-Hellman Key Exchange Implementation from Scratch

## Step 1: Introduction to Diffie-Hellman Key Exchange
# Diffie-Hellman key exchange is a method of securely exchanging cryptographic keys over a public channel.
# This protocol allows two parties to establish a shared secret over an insecure channel.

# Import necessary libraries
import random

## Step 2: Choose Parameters
# Choose a large prime number p and a base g.

# For simplicity, we'll use smaller values. In practice, these should be very large primes.
p = 23  # a prime number
g = 5   # a primitive root modulo p

print(f"Publicly known parameters:\np (prime number): {p}\ng (base): {g}")

## Step 3: Generate Private and Public Keys
# Each participant (Alice and Bob) generates a private key and a corresponding public key.

def generate_private_key(p):
    return random.randint(1, p - 1)

def generate_public_key(p, g, private_key):
    return pow(g, private_key, p)

# Alice's keys
alice_private_key = generate_private_key(p)
alice_public_key = generate_public_key(p, g, alice_private_key)

# Bob's keys
bob_private_key = generate_private_key(p)
bob_public_key = generate_public_key(p, g, bob_private_key)

print(f"Alice's public key: {alice_public_key}")
print(f"Bob's public key: {bob_public_key}")

## Step 4: Compute Shared Secret
# Both Alice and Bob compute the shared secret using their private key and the other's public key.

def compute_shared_secret(p, private_key, public_key):
    return pow(public_key, private_key, p)

alice_shared_secret = compute_shared_secret(p, alice_private_key, bob_public_key)
bob_shared_secret = compute_shared_secret(p, bob_private_key, alice_public_key)

print(f"Alice's shared secret: {alice_shared_secret}")
print(f"Bob's shared secret: {bob_shared_secret}")

# Both Alice and Bob now share the same secret, which can be used to encrypt further communication.
assert alice_shared_secret == bob_shared_secret, "Shared secrets do not match!"

```
